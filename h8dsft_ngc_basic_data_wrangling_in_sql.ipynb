```sql

-- Create table Customers and Orders

CREATE TABLE Customers (
	customer_id SERIAL PRIMARY KEY,
	customer_name TEXT NOT NULL,
	city VARCHAR(50) NOT NULL);

CREATE TABLE Orders (
	order_id SERIAL PRIMARY KEY,
	customer_id INTEGER NOT NULL,
	order_date DATE NOT NULL,
	total_amount FLOAT);

-- Melihat apakah table Customers dan Orders sudah terbentuk

SELECT *
FROM Customers;

SELECT *
FROM Orders;

-- Insert items kedalam table Customers dan Orders

INSERT INTO Customers (customer_id, customer_name, city)
VALUES
		(1, 'John Doe', 'New York'),
		(2, 'Jane Smith', 'Los Angeles'),
		(3, 'David Johnson', 'Chicago');

INSERT INTO Orders (order_id, customer_id, order_date, total_amount)
VALUES
		(1, 1, '2022-01-10', 100.00),
		(2, 1, '2022-02-15', 150.00),
		(3, 2, '2022-03-20', 200.00),
		(4, 3, '2022-04-25', 50.00);

```

# RESULT:






```sql

-- Menghitung total_order dari tiap customer dari tabel Customers dan Orders

SELECT Customers.customer_name, COUNT(Customers.customer_id) as total_order
FROM Customers
JOIN Orders ON Orders.customer_id = Customers.customer_id
GROUP BY Customers.customer_id
ORDER BY total_order DESC;

```
